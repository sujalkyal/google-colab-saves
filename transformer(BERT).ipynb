```python
from transformers import pipeline
classifier = pipeline('sentiment-analysis')
```

```python
classifier('i am very happy to get an internship')
```

```python
classifier('this sock smells too bad')
```

```python
results=classifier(
    ["this is awesome","you are dead","i am a bad person","he is the killer"]
)
for result in results:
  print(f"label: {result['label']} with score: {round(result['score'],4)}")
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
```

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
# This model only exists in PyTorch, so we use the `from_pt` flag to import that model in TensorFlow.
model = TFAutoModelForSequenceClassification.from_pretrained(model_name, from_pt=True)
tokenizer = AutoTokenizer.from_pretrained(model_name)
classifier = pipeline('sentiment-analysis', model=model, tokenizer=tokenizer)
```

```python
classifier("i am all good for now")
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
model_name = "distilbert-base-uncased-finetuned-sst-2-english"
tf_model = TFAutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
inputs = tokenizer("We are very happy to show you the 🤗 Transformers library.")
```

```python
print(inputs)
```

```python
tf_batch = tokenizer(
    ["this is awesome","you are dead","i am a bad person","he is the killer"],
    padding=True,
    truncation=True,
    max_length=512,
    return_tensors="tf"
)
```

```python
for key,value in tf_batch.items():
  print(f"{key}: {value.numpy().tolist()}")
```

```python
tf_outputs = tf_model(tf_batch)
```

```python
print(tf_outputs)
```

```python
## applying softmax activation function
import tensorflow as tf
tf_predictions = tf.nn.softmax(tf_outputs[0], axis=-1)
```

```python
print(tf_predictions)
```

```python
import tensorflow as tf
tf_outputs = tf_model(tf_batch, labels = tf.constant([1,0]))
```
